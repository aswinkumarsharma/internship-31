# 2.Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
and make data frame

```python
imdbpage=requests.get("https://www.imdb.com/search/title/?count=100&groups=top_100&sort=user_rating")
soup=BeautifulSoup(imdbpage.content,'html.parser')
imdbpage
```

```python
extract=soup.find_all('div',class_="lister-item mode-advanced")
name=[]
rating=[]
year=[]
for content in extract:
    movie=content.h3.a.text
    name.append(movie)
    
    years=content.h3.find('span' , class_="lister-item-year text-muted unbold").text
    year.append(years)
    
    ratings=content.find('div',class_="inline-block ratings-imdb-rating").text.replace('\n\n','').replace('\n','')
    rating.append(ratings)
imdb=pd.DataFrame(list(zip(name,rating,year)),columns=['Moviename','Ratings','Year'])
imdb
```

# 3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of
release) and make data frame.

```python
imdbpage1=requests.get('https://www.imdb.com/list/ls056092300/')
data=BeautifulSoup(imdbpage1.content)
extract=data.find_all('div',attrs={'class':"lister-item-content"})
```

```python
name=[]
rating=[]
release=[]
for content in extract:
    hindi=content.h3.a.text
    name.append(hindi)
    
    year=content.find('span',class_="lister-item-year text-muted unbold").text
    release.append(year)
    
    ratings=content.find('span',class_="ipl-rating-star__rating").text
    rating.append(ratings)    
hindiimdb=pd.DataFrame(list(zip(name,rating,release)),columns=['Indianmovie','Rating','YearofRelease'])
hindiimdb
```

# 4) Write s python program to display list of respected former presidents of India(i.e. Name , Term of office)

```python
prpage=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
soup=BeautifulSoup(prpage.content)
prlist=soup.find_all('div',class_='presidentListing')
```

```python
Name=[]
office=[]
for content in prlist:
    president=content.h3.text
    Name.append(president)
    
    term=content.p.text.replace('Term of Office:','')
    office.append(term)
President=pd.DataFrame(list(zip(Name,office)),columns=['Presidents_Name','Term_of_office'])
President    
```

# 5.a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
odipage1=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
soup=BeautifulSoup(odipage1.content)
team=[]
matches=[]
point=[]
rating=[]
```

```python
extract=soup.find_all('tr',class_="rankings-block__banner")
for content in extract:
    teams=content.find('span',class_="u-hide-phablet").text
    team.append(teams)
    
    match=content.find('td',class_="rankings-block__banner--matches").text
    matches.append(match)
    
    points=content.find('td',class_="rankings-block__banner--points").text
    point.append(points)
    
    ratings=content.find('td',class_="rankings-block__banner--rating u-text-right").text.strip().replace('\n','').replace('\n \n\n','')
    rating.append(ratings.strip())
    
```

```python
extract2=soup.find_all('tr',class_="table-body")
n=1
for content in extract2:
    if n<=9:
        teams=content.find('span',class_="u-hide-phablet").text
        team.append(teams)

        match=content.find('td',class_="table-body__cell u-center-text").text
        matches.append(match)

        points=content.find('td',class_="table-body__cell u-center-text").text
        point.append(points)

        ratings=content.find('td',class_="table-body__cell u-text-right rating").text
        rating.append(ratings)
        
        n+=1
```

```python
ODI_Team=pd.DataFrame(list(zip(team,matches,point,rating)),columns=['Team','matches','points','rating'])
ODI_Team
```

# 7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :

```python
latestnews=requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(latestnews.content)
extract=soup.find_all('div',class_='LatestNews-container')



headline=[]
time=[]
newslink=[]


i=0

for content in extract:
    if i<=200:
        headline.append(content.find('a',class_='LatestNews-headline').text)
        time.append(content.find('time',class_='LatestNews-timestamp').text)
        news=content.find('a',class_='LatestNews-headline')
        newslink.append(news['href'])
        i+=1

CNBCHeadlines=pd.DataFrame(list(zip(headline,time,newslink)),columns=('Headline','Time','Newslink')) 
CNBCHeadlines
    
```

# 8) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.
https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details

```python
artpage=requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")

soup=BeautifulSoup(artpage.content)
scrap=soup.find('ul',class_='sc-9zxyh7-0 ffmPq')
title=[]
authors=[]
pubdate=[]
url=[]
i=0
for content in scrap:
    if i<=50:
        title.append(content.find('h2',class_="sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR").text)
        authors.append(content.find('span',class_='sc-1w3fpd7-0 pgLAT').text)
        pubdate.append(content.find('span',class_="sc-1thf9ly-2 bKddwo").text)
        urls=content.find('a','sc-5smygv-0 nrDZj')
        url.append(urls['href'])

        i+=1

Art_page=pd.DataFrame(list(zip(title,authors,pubdate,url)),columns=('Papertitles','Authors','Publish_date','PaperURL')) 
Art_page
```

#    9) Write a python program to scrape mentioned details from dineout.co.in :

```python
dinepage=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup=BeautifulSoup(dinepage.content)
scrap=soup.find('div',class_='restnt-card-wrap-new')
Rname=[]
cuisine=[]
location=[]
ratings=[]
imageurl=[]

i=0
for content in scrap:
    if i<=50:
        Rname.append(content.find('a',class_='restnt-name ellipsis').text)
        
        cusi=content.find('span',class_='double-line-ellipsis').text
        length=len(cusi)
        cuisine.append(cusi[25:length])
        
        location.append(content.find('div',class_='restnt-loc ellipsis').text)
        
        ratings.append(content.find('div',class_='restnt-rating rating-4').text)
        
        img=content.find('img',class_='no-img')
        imageurl.append(img['data-src']) 
        
Dineout=pd.DataFrame(list(zip(Rname,cuisine,location,ratings,imageurl)),columns=('RestaurantName','Cuisine','Location','Ratings','ImageURL')) 
Dineout
```

# 10) Write a python program to scrape the details of top publications from Google Scholar from
https://scholar.google.com/citations?view_op=top_venues&hl=en

```python
page1=requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
soup=BeautifulSoup(page1.content)
rank=[]
publication=[]
index1=[]
index2=[]
for i in soup.find_all('td',class_="gsc_mvt_p"):
    rank.append(i.text)
for i in soup.find_all('td',class_='gsc_mvt_t'):
    publication.append(i.text)
for i in soup.find_all('td',class_='gsc_mvt_n'):
    index1.append(i.text)
for i in soup.find_all('span',class_='gs_ibl gsc_mp_anchor'):
    index2.append(i.text)
    
df=pd.DataFrame(list(zip(rank,publication,index1,index2)),columns=['rank','publication','h5-index','h5-median'])  
df
```
